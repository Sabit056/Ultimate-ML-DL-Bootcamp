### `Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Sho"]
list2 = ["y", "me", "s", "vo"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Sho', 'vo']]
```

````python
list1 = input("Enter: ").split()
list2 = input("Enter: ").split()

ls = [[a, b] for a, b in zip(list1, list2)]

if len(list1) > len(list2):
    for i in range(len(list2), len(list1)):
        ls.append([list1[i]])
elif len(list2) > len(list1):
    for i in range(len(list1), len(list2)):
        ls.append([list2[i]])

print(ls)
````

### `Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
ls = list(map(int,input().split()))
ans=[]
for i in range(len(ls)):
    sum=0
    for j in range(0,len(ls)):
        if ls[j]>=ls[i]:
            sum+=ls[j]
    ans.append(sum)
print(ans)
````

### `Q16:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
n = int(input())
test_list=[]
for _ in range(n):
    a,b = map(int,input().split())
    test_list.append((a,b))
res=[]
key=None
curr=[]
for a,b in test_list:
    if key!=a:
        if curr:
            res.append(tuple(curr))
        key=a
        curr=[a,b]
    else:
        curr.append(b)
res.append(tuple(curr))
print(res)
````

### `Q17:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
n=int(input())
a=[]
for _ in range(n):
    val=int(input())
    a.append(val)
t=tuple(a)
ans=[]
for i in range(len(t)):
    if i==0:
        s=t[i]*t[i+1]
    elif i==len(t)-1:
        s=t[i]*t[i-1]
    else:
        s=t[i-1]*t[i] + t[i]*t[i+1]
    ans.append(s)
print(tuple(ans))
````

### `Q26`: Key with maximum unique values

Given a dictionary with values list, extract key whose value has most unique values.

**Example 1:**

Input:

```bash
test_dict = {"CS" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "SHOVO" : [9, 9, 6, 5, 5]}
```

Output:
```bash
CS
```

**Example 2:**

Input:
```bash
test_dict = {"CS" : [5, 7, 7, 7, 7], "is" : [6, 7, 7, 7], "SHOVO" : [9, 9, 6, 5, 5]}
```

Output:
```bash
SHOVO
```

````python
data = {}
n = int(input("Enter number of entries: "))

for _ in range(n):
    key = input("key: ")
    val = [int(x) for x in input("value: ").split()]
    data[key] = val

rest = ""
mx = 0
for k, v in data.items():
    freq = {}
    for num in v:
        freq[num] = 1 
    
    cnt = sum(freq.values()) 
    
    if cnt > mx:
        mx = cnt
        rest = k
        
print(rest)
````
